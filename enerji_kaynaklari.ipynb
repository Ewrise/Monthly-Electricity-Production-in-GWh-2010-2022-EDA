### <span style="color:#00FF00">6 - Sonuç</span>

#### <span style="color:#00FF00">Özet</span>

##### Bu çalışmada, 2014 yılından sonra enerji kaynaklarının üretiminde önemli bir artış olduğu tespit edilmiştir. En çok enerji kaynağı Ocak ayında üretilmektedir ve fosil yakıtlar en çok tercih edilen enerji kaynağıdır, ardından kömür ve doğalgaz gelmektedir. Jeotermal enerji kaynakları en az kullanılan enerji kaynağıdır. ABD, Japonya, Hindistan, Almanya ve Kanada en çok enerji kaynağı üreten ülkelerdir. ABD ve Japonya'da açık ara farkla en çok fosil yakıt kullanılmaktadır ve fosil yakıttan sonra kömür ve doğalgaz elektrik üretimi için en çok kullanılan enerji kaynaklarıdır. Hindistan'da fosil yakıt ve kömür birbirine yakın değerlerle elektrik üretimi için en çok kullanılan enerji kaynakları arasındadır. Almanya'da fosil yakıt ve kömür birbirine yakın değerlerle elektrik üretimi için en çok kullanılan enerji kaynaklarıdır ve aynı zamanda doğalgaz, nükleer enerji ve rüzgar enerjisinden de oldukça yararlanılmaktadır. Kanada'da hidroelektrik enerjisi ülkenin enerji üretiminin yarısından fazlasını karşılamaktadır. Diğer yarısı da fosil yakıt, nükleer enerji, doğalgaz ve kömürden karşılanmaktadır. Geçen yıla göre bu yıl neredeyse bütün enerji kaynaklarının üretimi iki katına çıkmıştır. Bu bilgiler, enerji kaynakları hakkında farkındalığı artırmak için önemli bir kaynak olabilir.

#### <span style="color:#00FF00">Öneriler</span>

##### Yenilenebilir enerji kaynaklarının kullanımını arttırmak:
##### Bilindiği gibi fosil yakıtların kullanımı iklim değişikliği ve çevre kirliliği gibi sorunlara neden olmaktadır. Bu nedenle, enerji şirketi yenilenebilir enerji kaynaklarının kullanımını arttırmalıdır. Özellikle Almanya'da rüzgar enerjisinden oldukça yararlanılıyor, bu alanda yatırımların devam ettirilmesi ve diğer ülkelerde de bu alanda yatırımlar yapılması önerilebilir.

##### Enerji verimliliğini arttırmak:
##### Enerji şirketi, enerji verimliliğinin arttırılmasına yönelik projeleri desteklemelidir. Bu sayede hem enerji tüketimi azaltılabilir hem de maliyetler düşürülebilir. Örneğin, Kanada'da hidroelektrik enerjisinin kullanımı yüksek olsa da diğer yarısı fosil yakıt, nükleer enerji, doğalgaz ve kömürden karşılanıyor. Burada enerji verimliliği projeleri ile bu kaynakların kullanımı azaltılabilir.

##### Alternatif enerji kaynaklarına yatırım yapmak:
##### Enerji şirketi, alternatif enerji kaynaklarına yatırım yaparak çeşitlendirme yoluna gidebilir. Jeotermal enerji kaynakları, enerji kaynakları arasında en az kullanılanıdır. Ancak, özellikle ülkemizde bu alanda oldukça potansiyel olduğu düşünülmektedir. Bu nedenle, enerji şirketi jeotermal enerji kaynaklarına yatırım yaparak çeşitlendirme yoluna gidebilir.

##### Fosil yakıt kullanımını azaltmak:
##### Fosil yakıt kullanımı dünya genelinde büyük bir sorun haline gelmiştir. Enerji şirketi, fosil yakıt kullanımını azaltacak projelere destek olabilir veya kendi bünyesinde bu yönde çalışmalar yapabilir. Özellikle ABD ve Japonya gibi ülkelerde fosil yakıt kullanımı açık ara farkla en yüksek seviyededir, bu ülkelerde fosil yakıt kullanımının azaltılmasına yönelik çalışmalar yapılabilir.
